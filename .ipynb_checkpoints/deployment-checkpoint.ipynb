```python
btn_upload = widgets.FileUpload()
btn_upload
```

```python
img = PILImage.create(btn_upload.data[-1]) #btn_upload.data is an array of all images that were uploaded
img
```

```python
out_pl = widgets.Output()
out_pl.clear_output()
out_pl
```

```python
with out_pl: display(img.to_thumb(128,128))
```

```python
lbl_pred = widgets.Label()
lbl_pred.value = 'Choose an image'
lbl_pred
```

```python
btn_run = widgets.Button(description='Classify!')
btn_run
```

```python
def on_click_classify(change):
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(256,256))
    pred, pred_idx, probs = learn_inf.predict(img)
    lbl_pred.value = f'Prediction: {pred}: Probability: {probs[pred_idx]:.04f}'
    
btn_run.on_click(on_click_classify) #this is where we tell to run the function on click
```

Now our `Clasify!` button runs the code
